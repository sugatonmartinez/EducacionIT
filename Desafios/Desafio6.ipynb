# Ejercicio 1

Construir 3 gráficos de línea a partir de un array 
creado con la función linspace de numpy.
 - a. Uno que represente una curva lineal.
 - b. Otro que represente una curva cuadrática.
 - c. Un tercero que represente una curva logarítmica.

```python
import numpy as np
import matplotlib.pyplot as plt
# a: Grafico lineal
b,m = 10,2
x = np.linspace(-10,10,10)
y = m*x + b
plt.plot(x,y)
plt.title('Gráfico líneal')
```

```python
#b: Curva cuadratica
import random as rd
x = np.linspace(-100,100,100)
a = rd.randint(1,10)
b = rd.randint(1,10)
c = rd.randint(1,10)
y = a*x**2+b*x+c
print(f'La función de este gráfico es: {a}*x^2+{b}*x+{c}')
plt.plot(x,y)
plt.title('Cuadratica')
```

```python
#c: Gráfico logaritmico

x= np.linspace(1,10,100)
y = np.log(x)
plt.plot(y)
plt.title('Logaritmica')
```

# Ejercicio 3
###  Se tiene una lista de listas de la evolución del dólar por año y un diccionario con la demanda según en porcentaje de los que compran dólares o hacen uso de la divisa en Argentina:
### dolarValor = [[2020,70],[2021,95],[2022,110]]
### (Año – Valor dólar oficial)
### dolarComp = {“privados”:40,”particulares”:20,”estatales”:30,”otros”:10}
### Hacer un subplot con un gráfico de barras de dolarValor 
### y una torta de dolarComp con sus respectivos títulos. 


```python
dolarvalor = [[2020,70],[2021,95],[2022,110]]
dolarcomp =  {'privados':40,'particulares':20,'estatales':30,'otros':10}
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
arr_dv = np.array(dolarvalor)
plt.bar(arr_dv[:,0], arr_dv[:,1], label=[2020,2021,2022])
plt.title('Valor del dolar por año', size=20)
plt.subplot(1,2,2)
etiqueta = list(dolarcomp.keys())
valores = list(dolarcomp.values())
plt.pie(valores, labels=etiqueta,autopct='%1.f%%')
plt.title('Demanda del dolar', size=20)
```
